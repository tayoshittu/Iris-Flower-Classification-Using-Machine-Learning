#implement peceptron algorithin



```
#Data preparation
df=pd.read_csv('iris.data',header=None)

#shuffle dataset
df=df.sample(frac=1).reset_index(drop=True)
inputs=df[df.columns[0:4]]

#padding for matrix
inputs['clamped'] = 1

output=df[df.columns[4:5]]
setosa = (output == 'Iris-setosa').astype(int)
virginica = (output == 'Iris-virginica').astype(int)
versicolor = (output == 'Iris-versicolor').astype(int)

```

```
def data_preparation(inputs, output):
  train_size = int(0.8 * len(inputs))
  train_set_inputs =inputs[:train_size].values
  test_set_inputs = inputs[train_size:].values
  train_set_output = output[:train_size].values
  test_set_output= output[train_size:].values
  
  return train_set_inputs, train_set_output, test_set_inputs, test_set_output
```

The perceptron algorithm

Take wj random
START: Take X ε F+ U F

CHECK: if x ε F+ and Σ wjxj > 0 goto START

if x ε F+ and Σ wjxj ≤ 0 goto ADD

if x ε F- and Σ wjxj ≤ 0 goto START

if x ε F- and Σ wjxj > 0 goto SUB

ADD: wj → wj + xj
goto START

SUB: wj → wj - xj
goto START:

```
def train_perceptron(train_set_inputs, train_set_output, lr, num_of_epochs):
  weights = np.random.rand(5)
  print(weights)



  for epochs in range(num_of_epochs):
    print(epochs+1,"epoch:")
    
    for i, inp in enumerate(train_set_inputs):
      c=1 if inp.dot(weights) > 0 else 0

      if (c == train_set_output[i]):
        continue

      print("Step",i+1) #at what step in your training does your algo converge
      e=train_set_output[i]-c
      weights=weights+(e*lr*inp)   
      print("weights",weights) 

  return weights


def test_perceptron(weights,test_set_inputs,test_set_output):
  
  #Counter for correct predictions 
  
  corrects = 0
  
  for i, inp in enumerate(test_set_inputs):
    c=1 if inp.dot(weights) > 0 else 0
    
    if (c == test_set_output[i]):
      corrects = corrects + 1 
  
    else:
      continue
   
  
  accuracy = corrects/len(test_set_inputs) * 100
 
  return accuracy
```

```
#setosa vs non-setosa
features_train, label_train, features_test, label_test = data_preparation(inputs, setosa)
setosa_weights = train_perceptron(features_train, label_train, lr=1, num_of_epochs=5)
print(test_perceptron(setosa_weights, features_test, label_test))
```

```
#virginica vs non-virginica
features_train, label_train, features_test, label_test = data_preparation(inputs, virginica)
virginica_weights = train_perceptron(features_train, label_train, lr=0.01, num_of_epochs=10)
print(test_perceptron(virginica_weights, features_test, label_test))
```

```
#versicolor vs non-versicolor
features_train, label_train, features_test, label_test = data_preparation(inputs, versicolor)
versicolor_weights = train_perceptron(features_train, label_train, lr=0.01, num_of_epochs=10)
print(test_perceptron(versicolor_weights, features_test, label_test))
```

```
def nn_iris(nn_inputs):
  
  #Hidden layer
  set_hid = int(nn_inputs.dot(setosa_weights) > 0)
  vir_hid = int(nn_inputs.dot(virginica_weights) > 0)

  nn_hidden = np.array([set_hid, vir_hid, 1])

  #Output layer
  versicolor_out = int(nn_hidden.dot(np.array([-1, -1, 0.5])) > 0)
  setosa_out = int(nn_hidden.dot(np.array([1, 0, -0.5])) > 0)
  virginica_out = int(nn_hidden.dot(np.array([0, 1, -0.5])) > 0)

  return versicolor_out, setosa_out, virginica_out
```

```
sepal_length = 1
sepal_width = 2
petal_length = 3
petal_width = 2

nn_inputs = np.array([sepal_length, sepal_width, petal_length, petal_width, 1])

nn_iris(nn_inputs)
```

# Keras Framework Application

```
import tensorflow as tf
from tensorflow.keras import layers
```

```
inputs  = []
outputs = []

d = {}
d['Iris-setosa\n']     = [1., 0., 0.]
d['Iris-versicolor\n'] = [0., 1., 0.]
d['Iris-virginica\n']  = [0., 0., 1.]

with open('iris.data') as f:
    lines=f.readlines()
    for line in lines:
        items=line.split(',')
        if len(items) == 5:
            inp =  [float(x) for x in items[0:4] ]
            inputs.append(inp)
            out = d[items[4]]
            outputs.append(out)
            
print( len(inputs), 'input patterns', len(outputs), 'output patterns')
```

Now we build a network. A stack of layers is captured by the tf.keras.Sequential model.

```
import numpy as np
import random
ids=random.sample(range(0,len(inputs)), 100) # generate 100 random ids
train_in = []
train_out=[]
for id in ids:
    train_in.append(inputs[id])
    train_out.append(outputs[id])
train_inputs  = np.array(train_in)
train_outputs = np.array(train_out)

val_input =[]
val_output=[]
validation_ids = list(set(range(0,len(inputs))) - set(ids))
for val_id in validation_ids:
    val_input.append(inputs[val_id])
    val_output.append(outputs[val_id])
val_inputs  = np.array(val_input)
val_outputs = np.array(val_output)
```

```
model = tf.keras.Sequential()
# an mlp with a given number of input nodes. Four input nodes, three output nodes 
nr_hidden = 6
nr_out    = 3 
model.add(layers.Dense(nr_hidden, activation = 'relu'))
model.add(layers.Dense(nr_out,activation='softmax'))
model.compile(optimizer=tf.keras.optimizers.SGD(0.05),loss='categorical_crossentropy', metrics=['accuracy'])
```

Create training set and bring them into a numpy array form. Let's use a 100 patterns for training. Also, build a validation set.


Now train

```
log = model.fit(train_inputs,train_outputs, validation_data=(val_inputs, val_outputs), epochs=1000,batch_size=30)
```
